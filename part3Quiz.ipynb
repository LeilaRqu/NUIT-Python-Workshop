```python
dna_file = "DNAresults.csv"
with open(dna_file, "r") as f:
    for line in f.readlines(): #currently in list format #added index to remove header
        print(line)

with open(dna_file, "r") as f:
    dna_in = f.readlines()[1:] #currently in list format #added index to remove header

print(dna_in)
```

```python
#make an empty dictionary
dna_dict = {}
```

```python
#loop through list of DNA results
for line in dna_in:
  nobreak = line.rstrip('\n') #remove line breaks
  commasplit = nobreak.split(',') #separate into series of lists
  sampleid = commasplit[0] #isolate keys
  data = commasplit[1:] #isolate values
  dna_dict[sampleid] = data #assign keys and values to dictionary
```

```python
#answer:
dna_file = 'DNAresults.csv'
with open(dna_file, 'r') as f:
  raw_lines = f.readlines()

raw_lines
```

```python
#answer cont:
#make a list of dictionaries with header as keys, entries as values
data = []
for rline in raw_lines[1:]: #skip header, going to hard code keys instead of using actual header
  clean_line = rline.rstrip('\n')
  row_list = clean_line.split(',') #get list that contains individual values (list of strings)
  row_dict = {
      'well': row_list[0],
      'sample': row_list[1],
      'site' : row_list[2],
      'result': row_list[3]}
  data.append(row_dict)
```
